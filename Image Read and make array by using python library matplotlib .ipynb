# Matplotlib Imshow - Read & Show Image Using Imread() &    " plt.imshow()" make array


If you worry about, how to read and show an image using the matplotlib library then here you will get a solution for your problem. Along with that, you will be got a bonus. The matplotlib imshow() function helps to show the image.
<br><br>
But plt.imshow() didnâ€™t work without mpimg.imread() function which is belongs to matplotlib.image module. So lets start practical.

```python
# Import Libraries
import matplotlib.pyplot as plt
import matplotlib.image as mpimg # image module for image reading
```

## Reading Image
Here, we use "mpimg.imread()" method. <br>
Which belongs to the matplotlib image module.

```python
img = mpimg.imread('Anshu.jpg') # give addres of image location
print(img)
```

Above you can see, when we call img then got numpy array but how. <br>
because<b><i> mpimg.imread() </b></i>function read <b>Anshu.jpg</b> image and store in <b><i>numpy array</b></i> format.

```python
print('Data type of image : ',type(img))
print('Shape of image : ',img.shape)
print('Dimention of image : ',img.ndim)
```

## To show an image, use <i> plt.imshow()</i> function

```python
plt.imshow(img)
plt.show()
```

### Now, removing the axis and increase figure size and then show teh same image

```python
plt.figure(figsize=(16,9))
plt.axis('on')
plt.imshow(img)
plt.show()
```

### Now, it looks great but can we add it's a color bar.<br>Yes, using <i> "plt.colorbar()" </i> function

```python
plt.figure(figsize=(16,9))
plt.axis("on")
plt.imshow(img)
plt.colorbar() # Show color bar of above image
plt.show()
```

Letâ€™s play with plt.imshow() functions parameter. Here use cmap means color map to show a single-channel image in a different color.

```python
single_channel = img[:,:,1] # get single channel data from img
plt.figure(figsize=(16,9))
plt.axis("on")
plt.imshow(single_channel, cmap = "nipy_spectral_r") # show image with hot color map
plt.colorbar()
plt.show()
```

```python
img3 = mpimg.imread('selfie.jpg')
single_channel = img3[:,:,1] # get single channel data from img
plt.figure(figsize=(16,9))
plt.axis("on")
plt.imshow(single_channel, cmap = "RdYlGn") # show image with hot color map
plt.colorbar()
plt.show()
```

# Here are I am using OOPs concepte 
# I'll make arrary of image and read image

```python
class Imshow:
    def __init__(self,img_address,color):
        self.img_address=img_address
        self.color=color
    def plot_Imshow(self):
        import matplotlib.pyplot as plt
        import matplotlib.image as mpimg
        img = mpimg.imread(self.img_address )
        single_channel = img[:,:,1] # get single channel data from img
        plt.figure(figsize=(16,9))
        plt.axis("on")
        plt.imshow(single_channel, cmap = self.color) # show image with hot color map
        plt.colorbar()
        plt.show()
        print('array of image : ',img)
        
    
```

```python
#Imshow(image name , color name)
img=Imshow('bird.jpg','nipy_spectral')
img.plot_Imshow()
```
